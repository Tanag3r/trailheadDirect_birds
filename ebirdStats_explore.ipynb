```python
#All observations in 2020 and 2021 from April to September
date_range = []
for i in range(2020,2021):
    days = pd.date_range(start=dt.datetime(i,4,1), end=dt.datetime(i,9,1))
    date_range.append(days)
```

```python
date_list = []
for sublist in date_range:
        for item in sublist:
            date_list.append(item)
```

```python
token = 'j6c7l80ga2ib'
```

```python
frames = []

for date in date_list:
    url = '''https://api.ebird.org/v2/data/obs/US-WA-003/historic/{}/{}/{}'''.format(date.year, date.month, date.day)
    payload = {}
    headers = {'X-eBirdApiToken': token}
    response = requests.request("GET", url, headers=headers,data= payload)
    frame = pd.DataFrame(response.json())
    frames.append(frame)
```

```python
all_USWA003_data = pd.concat(frames)
```
